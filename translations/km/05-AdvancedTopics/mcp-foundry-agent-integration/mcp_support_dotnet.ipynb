# អ្នកទាយ Azure AI ជាមួយនឹងគម្រោងបែបបទបរិបទម៉ូដែល (MCP)

## អត្ថប្រយោជន៍នៃការផ្ទៀងផ្ទាត់គ្មានកូនសោ

សៀវភៅបញ្ជាក់នេះបង្ហាញពី **ការផ្ទៀងផ្ទាត់គ្មានកូនសោ** ដែលផ្តល់អត្ថប្រយោជន៍ជាច្រើន៖
- ✅ **គ្មានកូនសោ API ត្រូវគ្រប់គ្រង** - ប្រើការផ្ទៀងផ្ទាត់ផ្អែកលើអត្តសញ្ញាណ Azure
- ✅ **សុវត្ថិភាពកើនឡើង** - គ្មានសម្ងាត់ទុកនៅក្នុងកូដ ឬការកំណត់រចនាសម្ព័ន្ធ
- ✅ **ការបង្រួមបញ្ច្រាសគោលបំណងដោយស្វ័យប្រវត្តិ** - Azure គ្រប់គ្រងវដ្ដជីវិតគោលបំណង
- ✅ **ការចូលប្រើតាមតួនាទី** - ប្រើ Azure RBAC សម្រាប់សិទ្ធិលម្អិត

`DefaultAzureCredential` នឹងប្រើប្រភពគោលបំណងដែលមានលក្ខណៈល្អបំផុតដោយស្វ័យប្រវត្តិ៖

2. គោលបំណង Azure CLI (ក្នុងអំឡុងពេលអភិវឌ្ឍន៍)
3. គោលបំណង Visual Studio
4. អថេរសេវាពិតប្រាកដ (បើបានកំណត់)

## កំណត់រចនាសម្ព័ន្ធ Azure AI Agent Client (ការផ្ទៀងផ្ទាត់គ្មាន Key)

កំណត់អថេររចនាសម្ព័ន្ធ និងបង្កើត PersistentAgentsClient ដោយប្រើ **ការផ្ទៀងផ្ទាត់គ្មាន key**៖
- **projectEndpoint**: ចំណុចបញ្ចប់គម្រោង Azure AI Foundry
- **modelDeploymentName**: ឈ្មោះម៉ូដែល AI ដែលបានដាក់បង្ហាញ (GPT-4.1 nano)
- **mcpServerUrl**: URL នៃម៉ាស៊ីនមេ MCP (Microsoft Learn API)
- **mcpServerLabel**: ស្លាកសម្រាប់កំណត់ម៉ាស៊ីនមេ MCP
- **DefaultAzureCredential**: ប្រើ managed identity, Azure CLI, ឬប្រភពសញ្ញាប័ត្រ​ផ្សេងទៀត (មិនត្រូវការ key API)


```polyglot-notebook
var projectEndpoint = "Your Azure AI Foundry Project Endpoint";
var modelDeploymentName = "Your Azure OpenAI Model Deployment Name";
var mcpServerUrl = "https://learn.microsoft.com/api/mcp";
var mcpServerLabel = "mslearn";
PersistentAgentsClient agentClient = new(projectEndpoint, new DefaultAzureCredential());
```

## បង្កើតនិយមន័យឧបករណ៍ MCP

បង្កើតនិយមន័យឧបករណ៍ MCP ដែលភ្ជាប់ទៅម៉ាស៊ីនមេ Microsoft Learn MCP។ វានឹងអនុញ្ញាតឲ្យភ្នាក់ងារចូលប្រើមាតិកានិងឯកសាររបស់ Microsoft Learn។


```polyglot-notebook
MCPToolDefinition mcpTool = new(mcpServerLabel, mcpServerUrl);
```

## បង្កើតភ្នាក់ងារ AI

បង្កើតភ្នាក់ងារ AI មានភាពអចិន្ត្រៃ្យជាមួយម៉ូដែល និងឧបករណ៍ MCP ដែលបានបញ្ជាក់។ ភ្នាក់ងារត្រូវបានកំណត់ជាមួយ៖
- ម៉ូដែល GPT-4.1 nano
- សេចក្ដីណែនាំក្នុងការប្រើឧបករណ៍ MCP សម្រាប់ជំនួយ
- ការចូលដំណើរការ​ម៉ាស៊ីនមេ Microsoft Learn MCP


```polyglot-notebook
PersistentAgent agent = await agentClient.Administration.CreateAgentAsync(
   model: modelDeploymentName,
   name: "my-learn-agent",
   instructions: "You are a helpful agent that can use MCP tools to assist users. Use the available MCP tools to answer questions and perform tasks.",
   tools: [mcpTool]
   );
```

## បង្កើតសេចក្តីពិភាក្សា និងផ្ញើសារ

បង្កើតខ្សែសន្ទនាដើម្បីផ្ញើសារមួយទៅអ្នកប្រើប្រាស់ដែលសួរអំពីភាពខុសគ្នារវាង Azure OpenAI និង OpenAI។ នេះនឹងសាកល្បងសមត្ថភាពរបស់ភ្នាក់ងារជាអ្នកប្រើឧបករណ៍ MCP ដើម្បីផ្តល់ព័ត៌មានដែលត្រឹមត្រូវ។


```polyglot-notebook
PersistentAgentThread thread = await agentClient.Threads.CreateThreadAsync();

// Create message to thread
PersistentThreadMessage message = await agentClient.Messages.CreateMessageAsync(
    thread.Id,
    MessageRole.User,
    "What's difference between Azure OpenAI and OpenAI?");
```

## កំណត់ធនធានឧបករណ៍ MCP (គ្មានកូនសោ)

តំឡើងធនធានឧបករណ៍ MCP។ សម្រាប់វិធីសាស្រ្តគ្មានកូនសោពិតប្រាកដ អ្នកអាចលុបចេញផ្នែកស៊ែដរប័ន្តផ្ទុកប្ដូរបញ្ចូលបើម៉ាស៊ីនបម្រើ MCP គាំទ្រការផ្ទៀងផ្ទាត់ខ្លួនលើមូលដ្ឋានអតិថិជនអាសឪរងรស៊ី។ ឧទាហរណ៍ខាងក្រោមបង្ហាញពីរបៀបបន្ថែមស៊ែដរប័ន្តបើត្រូវការ ប៉ុន្តាសម្រាប់ស្ថានភាពគ្មានកូនសោ អាចមិនចាំបាច់ត្រូវការ។


```polyglot-notebook
// Option 1: Completely keyless (if MCP server supports Azure identity)
MCPToolResource mcpToolResource = new(mcpServerLabel);
ToolResources toolResources = mcpToolResource.ToToolResources();

// Option 2: With custom headers (if still needed for specific MCP servers)
// MCPToolResource mcpToolResource = new(mcpServerLabel);
// mcpToolResource.UpdateHeader("Authorization", "Bearer <your-token>");
// ToolResources toolResources = mcpToolResource.ToToolResources();
```

## ចាប់ផ្តើមរត់ភូត

បង្កើត និងចាប់ផ្តើមរត់ដើម្បីដំណើរការចោលសាររបស់អ្នកប្រើ។ ភូតនឹងប្រើឧបករណ៍ MCP និងធនធានដែលបានកំណត់រចនាឡើងដើម្បីបង្កើតចម្លើយ។


```polyglot-notebook
ThreadRun run = await agentClient.Runs.CreateRunAsync(thread, agent, toolResources);
```

## ត្រួតពិនិត្យការប្រតិបត្តិរបស់កម្មវិធី និងដោះស្រាយការអនុម័តឧបករណ៍ (គ្មានសោ)

ត្រួតពិនិត្យស្ថានភាពការប្រតិបត្តិរបស់ភ្នាក់ងារ និងដោះស្រាយការអនុម័តឧបករណ៍ណាមួយដែលត្រូវការ។ ខ្សែរកតនេះ៖
1. រង់ចាំឲ្យការប្រតិបត្តិបានបញ្ចប់ ឬត្រូវការជំហាន
2. អនុម័តការហៅឧបករណ៍ MCP ដោយស្វ័យប្រវត្តិ នៅពេលដែលត្រូវការ
3. សម្រាប់ការផ្ទៀងផ្ទាត់គ្មានសោ មិនចាំបាច់ត្រូវការ header ប្រសិនបើម៉ាស៊ីនបម្រើ MCP គាំទ្រអត្តសញ្ញាណ Azure


```polyglot-notebook
while (run.Status == RunStatus.Queued || run.Status == RunStatus.InProgress || run.Status == RunStatus.RequiresAction)
{
    await Task.Delay(TimeSpan.FromMilliseconds(1000));
    run = await agentClient.Runs.GetRunAsync(thread.Id, run.Id);

    if (run.Status == RunStatus.RequiresAction && run.RequiredAction is SubmitToolApprovalAction toolApprovalAction)
    {
        var toolApprovals = new List<ToolApproval>();
        foreach (var toolCall in toolApprovalAction.SubmitToolApproval.ToolCalls)
        {
            if (toolCall is RequiredMcpToolCall mcpToolCall)
            {
                Console.WriteLine($"Approving MCP tool call: {mcpToolCall.Name}");
                
                // Option 1: Keyless approval (no headers needed)
                toolApprovals.Add(new ToolApproval(mcpToolCall.Id, approve: true));
                
                // Option 2: With headers (if required by specific MCP server)
                // toolApprovals.Add(new ToolApproval(mcpToolCall.Id, approve: true)
                // {
                //     Headers = { ["Authorization"] = "Bearer <your-token>" }
                // });
            }
        }

        if (toolApprovals.Count > 0)
        {
            run = await agentClient.Runs.SubmitToolOutputsToRunAsync(thread.Id, run.Id, toolApprovals: toolApprovals);
        }
    }
}
```

## បង្ហាញលទ្ធផលការសន្ទនា

ទាញយក និងបង្ហាញសារទាំងអស់នៅក្នុងខ្សែសន្ទនា បង្ហាញទាំងសំណួររបស់អ្នកប្រើ និងចម្លើយរបស់ភ្នាក់ងារ។ សារនីមួយៗត្រូវបានបង្ហាញតាមលំដាប់ពេលវេលា ដោយមានម៉ោងនិងសញ្ញាប្រភេទតួអង្គ។


```polyglot-notebook
using Azure;

AsyncPageable<PersistentThreadMessage> messages = agentClient.Messages.GetMessagesAsync(
    threadId: thread.Id,
    order: ListSortOrder.Ascending
);

await foreach (PersistentThreadMessage threadMessage in messages)
{
    Console.Write($"{threadMessage.CreatedAt:yyyy-MM-dd HH:mm:ss} - {threadMessage.Role,10}: ");
    foreach (MessageContent contentItem in threadMessage.ContentItems)
    {
        if (contentItem is MessageTextContent textItem)
        {
            Console.Write(textItem.Text);
        }
        else if (contentItem is MessageImageFileContent imageFileItem)
        {
            Console.Write($"<image from ID: {imageFileItem.FileId}>");
        }
        Console.WriteLine();
    }
}
```
